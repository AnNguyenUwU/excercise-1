```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

Distance = ctrl.Antecedent(np.arange(0, 50.1, 0.1), 'Distance')
Traffic = ctrl.Antecedent(np.arange(0, 101, 1), 'Traffic')
Demand = ctrl.Antecedent(np.arange(0, 101, 1), 'Demand')
Weather = ctrl.Antecedent(np.arange(1, 5, 0.1), 'Weather')
Rating = ctrl.Antecedent(np.arange(1, 5.1, 0.1), 'Rating')
RidePunctuality = ctrl.Antecedent(np.arange(0, 101, 1), 'RidePunctuality')

Cost = ctrl.Consequent(np.arange(12,200.1,4.2), 'Cost')
Bonus = ctrl.Consequent(np.arange(0,11,1), 'Bonus')

Distance['S']=fuzz.trimf(Distance.universe,[0,0,3])
Distance['M']=fuzz.trimf(Distance.universe,[2,3,8])
Distance['L']=fuzz.trimf(Distance.universe,[6,8,20])
Distance['VL']=fuzz.trimf(Distance.universe,[15,20,50])

Traffic['L']=fuzz.trimf(Traffic.universe,[0,0,30])
Traffic['M']=fuzz.trimf(Traffic.universe,[20,30,70])
Traffic['H']=fuzz.trimf(Traffic.universe,[60,70,100])

Demand['L']=fuzz.trimf(Demand.universe,[0,0,30])
Demand['M']=fuzz.trimf(Demand.universe,[20,30,70])
Demand['H']=fuzz.trimf(Demand.universe,[60,70,100])

Weather['B']=fuzz.trimf(Weather.universe,[1,1,2.5])
Weather['M']=fuzz.trimf(Weather.universe,[2,2.5,4])
Weather['G']=fuzz.trimf(Weather.universe,[3.5,4,5])

Rating['P']=fuzz.trimf(Rating.universe,[1,1,2.5])
Rating['A']=fuzz.trimf(Rating.universe,[2,2.5,4])
Rating['G']=fuzz.trimf(Rating.universe,[3.5,4,5])

RidePunctuality['L']=fuzz.trimf(RidePunctuality.universe,[0,0,50])
RidePunctuality['OT']=fuzz.trimf(RidePunctuality.universe,[40,50,80])
RidePunctuality['E']=fuzz.trimf(RidePunctuality.universe,[70,80,100])

Cost['L']=fuzz.trimf(Cost.universe,[12,12,16.2])
Cost['M']=fuzz.trimf(Cost.universe,[12,16.2,45.6])
Cost['H']=fuzz.trimf(Cost.universe,[37.2,45.6,96])
Cost['VH']=fuzz.trimf(Cost.universe,[75,96,222])

Bonus['N']=fuzz.trimf(Bonus.universe,[0,0,0])
Bonus['F']=fuzz.trimf(Bonus.universe,[1,1,4])
Bonus['M']=fuzz.trimf(Bonus.universe,[2,4,6])
Bonus['H']=fuzz.trimf(Bonus.universe,[5,6,10])

rule1 = ctrl.Rule(Distance['S'] & Traffic['L'] & Demand['L'], Cost['L'])
rule2 = ctrl.Rule(Distance['S'] & Traffic['M'] & Demand['H'], Cost['M'])
rule3 = ctrl.Rule(Distance['M'] & Traffic['H'] & Demand['H'], Cost['H'])
rule4 = ctrl.Rule(Distance['L'] & Traffic['M'] & Weather['G'], Cost['M'])
rule5 = ctrl.Rule(Distance['L'] & Traffic['H'] & Weather['B'], Cost['VH'])
rule6 = ctrl.Rule(Distance['VL'] & Traffic['H'] & Demand['H'], Cost['VH'])
rule7 = ctrl.Rule(Distance['M'] & Traffic['L'] & Demand['L'], Cost['M'])
rule8 = ctrl.Rule(Distance['S'] & Traffic['H'] & Weather['B'], Cost['H'])
rule9 = ctrl.Rule(Distance['VL'] &  Weather['B'], Cost['VH'])
rule10 = ctrl.Rule(Distance['M'] & Traffic['M'] & Weather['M'], Cost['M'])
rule11 = ctrl.Rule(Rating['G'] & RidePunctuality['E'], Bonus['H'])
rule12 = ctrl.Rule(Rating['A'] & RidePunctuality['OT'], Bonus['M'])
rule13 = ctrl.Rule(Rating['P'] & RidePunctuality['L'], Bonus['N'])
rule14 = ctrl.Rule(Distance['L'] & Traffic['H'] & RidePunctuality['OT'], Bonus['H'])
rule15 = ctrl.Rule(Distance['M'] & Traffic['M'] & Rating['G'], Bonus['M'])
rule16 = ctrl.Rule(Distance['VL'] & Traffic['H'] & Rating['G'], Bonus['H'])
rule17 = ctrl.Rule(Distance['VL'] & Weather['B'] & Rating['G'], Bonus['H'])
rule18 = ctrl.Rule(Distance['S'] & Rating['A'] & RidePunctuality['OT'], Bonus['F'])
rule19 = ctrl.Rule(Distance['L'] & Traffic['H'] & RidePunctuality['L'], Bonus['F'])
rule20 = ctrl.Rule(Distance['M'] & Weather['M'] & Rating['G'], Bonus['M'])

biking_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,
                                  rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20])
biking_system = ctrl.ControlSystemSimulation(biking_ctrl)

biking_system.input['Distance'] = 40
biking_system.input['Traffic'] = 50
biking_system.input['Demand'] = 50
biking_system.input['Weather'] = 1
biking_system.input['Rating'] = 4
biking_system.input['RidePunctuality'] = 80

biking_system.compute()
print(biking_system.output['Cost'])
print(biking_system.output['Bonus'])
Cost.view(sim=biking_system)
Bonus.view(sim=biking_system)
```

```python
from typing_extensions import TypeAliasType
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

Rating = ctrl.Antecedent(np.arange(1, 5.1, 0.1), 'Rating')
Volume = ctrl.Antecedent(np.arange(0, 100.1, 0.1), 'Volume')
Profit = ctrl.Antecedent(np.arange(0, 100.1, 0.1), 'Profit')
Event = ctrl.Antecedent(np.arange(0, 101, 1), 'Event')
ComDiscounts = ctrl.Antecedent(np.arange(0, 101, 1), 'ComDiscounts')
Discounts = ctrl.Consequent(np.arange(0,70,1), 'Discounts')

Rating['L']=fuzz.trimf(Rating.universe,[0,0,4])
Rating['M']=fuzz.trimf(Rating.universe,[4,4.3,4.5])
Rating['H']=fuzz.trimf(Rating.universe,[4.5,5,5])

Volume['L']=fuzz.trimf(Volume.universe,[0,0,30])
Volume['M']=fuzz.trimf(Volume.universe,[30,50,70])
Volume['H']=fuzz.trimf(Volume.universe,[70,100,100])

Profit['L']=fuzz.trimf(Profit.universe,[0,0,30])
Profit['M']=fuzz.trimf(Profit.universe,[30,50,70])
Profit['H']=fuzz.trimf(Profit.universe,[70,100,100])

Event['N']=fuzz.trimf(Event.universe,[0,0,10])
Event['M']=fuzz.trimf(Event.universe,[10,40,60])
Event['H']=fuzz.trimf(Event.universe,[60,80,100])

ComDiscounts['L']=fuzz.trimf(ComDiscounts.universe,[0,0,30])
ComDiscounts['M']=fuzz.trimf(ComDiscounts.universe,[30,50,70])
ComDiscounts['H']=fuzz.trimf(ComDiscounts.universe,[70,100,100])

Discounts['VL']=fuzz.trimf(Discounts.universe,[0,3,5])
Discounts['L']=fuzz.trimf(Discounts.universe,[5,8,10])
Discounts['M']=fuzz.trimf(Discounts.universe,[10,15,20])
Discounts['H']=fuzz.trimf(Discounts.universe,[20,30,40])
Discounts['VH']=fuzz.trimf(Discounts.universe,[40,55,70])

rule1 = ctrl.Rule(Rating['H'] & Volume['H'] & Profit['H'], Discounts['VL'])
rule2 = ctrl.Rule(Rating['L'] & Volume['L'] & Profit['H'], Discounts['H'])
rule3 = ctrl.Rule(Event['H'] & ComDiscounts['H'], Discounts['VH'])
rule4 = ctrl.Rule(Rating['M'] & Volume['M'] & Profit['M'], Discounts['M'])
rule5 = ctrl.Rule(Volume['H'] & Profit['L'] & ComDiscounts['L'], Discounts['VL'])
rule6 = ctrl.Rule(Event['N'] & Rating['L'], Discounts['M'])
rule7 = ctrl.Rule(Volume['L'] & Profit['L'], Discounts['VL'])

shopping_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7])
shopping_system = ctrl.ControlSystemSimulation(shopping_ctrl)

shopping_system.input['Rating'] = 4.3
shopping_system.input['Volume'] = 50
shopping_system.input['Profit'] = 15
shopping_system.input['Event'] = 90
shopping_system.input['ComDiscounts'] = 85

shopping_system.compute()
print(shopping_system.output['Discounts'])
Discounts.view(sim=shopping_system)
```

```python
from typing_extensions import TypeAliasType
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

Density = ctrl.Antecedent(np.arange(0, 101, 1), 'Density')
Urgency = ctrl.Antecedent(np.arange(0, 11, 1), 'Urgency')
Profit = ctrl.Antecedent(np.arange(0, 100.1, 0.1), 'Profit')
Load = ctrl.Antecedent(np.arange(0, 101, 1), 'Load')
Traffic = ctrl.Antecedent(np.arange(0, 11, 1), 'Traffic')

Priority = ctrl.Consequent(np.arange(0,101,1), 'Priority')
Orders = ctrl.Consequent(np.arange(0,101,1), 'Orders')

Density['L']=fuzz.trimf(Density.universe,[0,0,30])
Density['M']=fuzz.trimf(Density.universe,[30,50,70])
Density['H']=fuzz.trimf(Density.universe,[70,100,100])

Urgency['L']=fuzz.trimf(Urgency.universe,[0,0,3])
Urgency['M']=fuzz.trimf(Urgency.universe,[2,5,8])
Urgency['H']=fuzz.trimf(Urgency.universe,[7,10,10])

Profit['L']=fuzz.trimf(Profit.universe,[0,0,30])
Profit['M']=fuzz.trimf(Profit.universe,[30,50,70])
Profit['H']=fuzz.trimf(Profit.universe,[70,100,100])

Load['L']=fuzz.trimf(Load.universe,[0,0,30])
Load['M']=fuzz.trimf(Load.universe,[30,50,70])
Load['H']=fuzz.trimf(Load.universe,[70,100,100])

Traffic['L']=fuzz.trimf(Traffic.universe,[0,0,3])
Traffic['M']=fuzz.trimf(Traffic.universe,[2,5,8])
Traffic['H']=fuzz.trimf(Traffic.universe,[7,10,10])

Priority['L']=fuzz.trimf(Priority.universe,[0,0,30])
Priority['M']=fuzz.trimf(Priority.universe,[30,50,70])
Priority['H']=fuzz.trimf(Priority.universe,[70,100,100])

Orders['L']=fuzz.trimf(Orders.universe,[0,0,30])
Orders['M']=fuzz.trimf(Orders.universe,[30,50,70])
Orders['H']=fuzz.trimf(Orders.universe,[70,100,100])

rule1 = ctrl.Rule(Density['H'] & Traffic['L'] & Load['L'], Orders['H'])
rule2 = ctrl.Rule(Density['M'] & Traffic['H'] & Urgency['M'], Orders['L'])
rule3 = ctrl.Rule(Density['H'] & Profit['M'] & Load['H'], Orders['M'])
rule4 = ctrl.Rule(Traffic['M'] & Density['L'] & Urgency['H'], Orders['L'])
rule5 = ctrl.Rule(Urgency['H'] & Profit['H'] & Traffic['H'], Orders['L'])
rule6 = ctrl.Rule(Urgency['H'] & Profit['H'], Priority['H'])
rule7 = ctrl.Rule(Traffic['M'] & Urgency['M'], Priority['M'])
rule8 = ctrl.Rule(Traffic['L'] & Urgency['M'], Priority['M'])
rule9 = ctrl.Rule(Urgency['L'] & Profit['L'] & Density['H'], Priority['L'])

shipping_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
shipping_system = ctrl.ControlSystemSimulation(shipping_ctrl)

shipping_system.input['Density'] = 80
shipping_system.input['Urgency'] = 6
shipping_system.input['Profit'] = 55
shipping_system.input['Load'] = 20
shipping_system.input['Traffic'] = 2

shipping_system.compute()
print(shipping_system.output['Orders'])
print(shipping_system.output['Priority'])
Orders.view(sim=shipping_system)
Priority.view(sim=shipping_system)
```
