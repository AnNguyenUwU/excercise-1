```python
import osmnx as ox
import matplotlib.pyplot as plt
import folium
import numpy as np

center_point = (10.7610288, 106.6682)
dist = 1000

graph = ox.graph_from_point(center_point, dist=dist, network_type='walk')
gdf_nodes, gdf_edges = ox.graph_to_gdfs(graph)

m_network = folium.Map(location=[center_point[0], center_point[1]], zoom_start=15)

folium.GeoJson(
    gdf_edges[['geometry', 'name', 'length']],
    name='Mạng lưới đường bộ',
    style_function=lambda x: {'color': 'blue', 'weight': 2, 'opacity': 0.6}
).add_to(m_network)

# Calculate area in square meters (circle area: pi * r^2)
area = np.pi * dist**2
stats = ox.basic_stats(graph, area=area)

print(f"--- THÔNG SỐ MẠNG LƯỚI ---")
print(f"Số lượng nút giao (Nodes): {stats['n']}")
print(f"Số lượng đoạn đường (Edges): {stats['m']}")
print(f"Chiều dài đường trung bình: {stats['edge_length_avg']:.2f} mét")
print(f"Mật độ nút giao: {stats['node_density_km']:.2f} nút/km²")
print(f"Mật độ mạng lưới đường: {stats['street_density_km']:.2f} mét/km²")

display(m_network)
```

```python
import folium
from geopy.distance import geodesic
import pandas as pd

data = pd.read_csv('/Bài 23.3.csv')
center_coords = [10.7610288, 106.6682]
m_service = folium.Map(location=center_coords, zoom_start=16)

service_zones = [
    {'radius': 200, 'color': 'green', 'label': 'Vùng ưu tiên (<200m)'},
    {'radius': 500, 'color': 'orange', 'label': 'Vùng tiêu chuẩn (200-500m)'},
    {'radius': 800, 'color': 'red', 'label': 'Vùng mở rộng (500-800m)'}
]
for zone in service_zones:
    folium.Circle(
        location=center_coords,
        radius=zone['radius'],
        color=zone['color'],
        fill=True,
        fill_opacity=0.1,
        tooltip=zone['label']
    ).add_to(m_service)

stats = {'Priority': 0, 'Standard': 0, 'Extended': 0}
for index, row in data.iterrows():
    point = [row['Latitude'], row['Longitude']]
    dist = geodesic(center_coords, point).meters

    if dist <= 200:
        zone_color = 'green'
        stats['Priority'] += row['Number']
    elif dist <= 500:
        zone_color = 'orange'
        stats['Standard'] += row['Number']
    else:
        zone_color = 'red'
        stats['Extended'] += row['Number']

    folium.Marker(
        location=point,
        popup=f"{row['Location']}: {dist:.1f}m",
        icon=folium.Icon(color=zone_color, icon='info-sign')
    ).add_to(m_service)

folium.Marker(
    location=center_coords,
    popup="TRUNG TÂM PHÂN PHỐI (UEH)",
    icon=folium.Icon(color='black', icon='home')
).add_to(m_service)

display(m_service)

print("--- THỐNG KÊ KHẢ NĂNG TIẾP CẬN ---")
print(f"Số lượng khách vùng Ưu tiên (<200m): {stats['Priority']}")
print(f"Số lượng khách vùng Tiêu chuẩn (200-500m): {stats['Standard']}")
print(f"Số lượng khách vùng Mở rộng (>500m): {stats['Extended']}")
```

### Nhận xét và Ý nghĩa Quản trị

1. **Nhận diện thị trường trọng điểm:** Bản đồ Choropleth giúp nhà quản lý xác định ngay lập tức các khu vực có màu đậm (độ tập trung đơn hàng cao). Đây là các thị trường cốt lõi cần được ưu tiên nguồn lực hạ tầng và kho bãi.
2. **Phát hiện vùng trũng:** Các khu vực màu nhạt cho thấy sự thâm nhập thị trường còn thấp. Quản trị viên cần đánh giá xem đây là do đặc thù địa lý, dân cư hay do thiếu hụt các chiến dịch marketing.
3. **Tối ưu hóa Logistics:** Dựa trên mật độ đơn hàng, doanh nghiệp có thể quyết định đặt thêm các trạm trung chuyển (Cross-docking) tại các tỉnh giáp ranh các vùng màu đậm để giảm thời gian và chi phí vận chuyển chặng cuối.

```python
import folium
from geopy.geocoders import Nominatim
import osmnx as ox
import networkx as nx

address_A = "Ben Thanh Market, Ho Chi Minh City, Vietnam"
address_B = "Tan Son Nhat Airport, Ho Chi Minh City, Vietnam"

geolocator = Nominatim(user_agent="routing_test", timeout=10)

location_A = geolocator.geocode(address_A)
location_B = geolocator.geocode(address_B)

if location_A and location_B:
    latlng_A = (location_A.latitude, location_A.longitude)
    latlng_B = (location_B.latitude, location_B.longitude)

    midpoint = ((latlng_A[0] + latlng_B[0]) / 2, (latlng_A[1] + latlng_B[1]) / 2)

    G = ox.graph_from_point(midpoint, dist=3000, network_type="drive")

    orig_node = ox.distance.nearest_nodes(G, latlng_A[1], latlng_A[0])
    dest_node = ox.distance.nearest_nodes(G, latlng_B[1], latlng_B[0])

    route = nx.shortest_path(G, orig_node, dest_node, weight='length')
    route_coords = [(G.nodes[n]['y'], G.nodes[n]['x']) for n in route]

    m = folium.Map(location=midpoint, zoom_start=14)

    folium.Marker(latlng_A, popup="Điểm A (Bến Thành)", icon=folium.Icon(color="red")).add_to(m)
    folium.Marker(latlng_B, popup="Điểm B (Tân Sơn Nhất)", icon=folium.Icon(color="blue")).add_to(m)

    folium.PolyLine(locations=route_coords, color='green', weight=5, opacity=0.8).add_to(m)

    display(m)
else:
    print("Không tìm thấy tọa độ cho một trong hai địa chỉ. Vui lòng kiểm tra lại tên địa chỉ.")
```
